```python
# create a dictionary with original name of league as key
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1',
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
for league in dict_countries:
  print(league)
```

```python
dict_historical_data={} #contain all dataframes i want

#create root variable
root ="https://www.football-data.co.uk/mmz4281/"

#create list of leagues
#leagues = ['E0','E1','E2','E3','EC']

#loop
for league in dict_countries:
  frames=[] #empty list
  for season in range (15,21): #range exclude last element:

    ddf=pd.read_csv(root+ str(season)+ str(season+1) +'/'+ dict_countries[league]+'.csv',encoding='unicode_escape')
    # to recognize year and season
    ddf.insert(1,'season',season)
    #ddf.insert()
    frames.append(ddf) #store eah dataframe in list
  ddf_concat=pd.concat(frames) #concate all seasons
  dict_historical_data[league]=ddf_concat
```
