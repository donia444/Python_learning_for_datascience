```python
dfp=pd.read_csv('/content/population_total.csv')
```

```python
dfp
```

```python
dfp.info()
```

```python
dfp=dfp.dropna()
```

```python
#reshape the data
dfp=dfp.pivot_table(index='year',columns='country',values='population')
```

```python
dfp
```

```python
dfp=dfp[["United States",	"India"	,"China"	,"Indonesia","Brazil"]]
```

```python
dfp.head(3)
```

```python
#LinePlot
dfp.plot(kind='line',xlabel='year',ylabel='population',title='population of countries',figsize=(8,4))
```

```python
dfp.index
```

```python
df_pivot_2020=dfp[dfp.index.isin([2020])]
```

```python
df_pivot_2020
```

```python
#transpose to switch columns and rows
df_pivot_2020=df_pivot_2020.T
```

```python
df_pivot_2020
```

```python
#BarPlot
df_pivot_2020.plot(kind='bar',color='Pink',xlabel='year',ylabel='population',title ='population_2020')
```

```python
dfp
```
