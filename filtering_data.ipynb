```python
#select method ,condition ,values
conditions=[
    df['Price_euros']>3000,
    (df['Price_euros']>2000) & (df['Price_euros']<=3000),
    (df['Price_euros']>1000) & (df['Price_euros']<=2000),
    (df['Price_euros']<=1000)

]

values = ['Too Expensive','Expensive','Affordable','Cheap']
```

```python
df["Price_tiers"]=np.select(conditions,values,default='Not Specified')
```

```python
df.head(2)
```

```python
df['Price_tiers'].value_counts()
```

```python
cond=[
    df["Inches"]>=16,
    (df["Inches"]<16) & (df["Inches"]>=14),
    (df["Inches"]<14) & (df["Inches"]>=12),
    df["Inches"]<12
]
val=["Too Big","Big","small","Tiny"]
```

```python
df["size_tier"]=np.select(cond,val,default="Not Specified")
```
