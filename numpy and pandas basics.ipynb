```python
#assign and insert method for adding column
score1=np.random.randint(1,100,size=395)
score2=np.random.randint(1,100,size=395)
```

```python
series1 = pd.Series(score1,index=np.arange(0,395))
series2 = pd.Series(score2,index=np.arange(0,395))
```

```python
df=df.assign(score1=series1,score2=series2)
df
```
